```python
from brits import BRITSimputation
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("traffic.csv",encoding='CP949')
df["time"] = pd.to_datetime(df["집계일시"],format="%Y%m%d%H%M")
df["value"] = df["평균속도"]
df = df[["time","value"]]

df.iloc[200:205,1] = np.nan
df.iloc[300:305,1] = np.nan
df.iloc[400:410,1] = np.nan
df.iloc[500:510,1] = np.nan
df.iloc[600:605,1] = np.nan
df.iloc[700:705,1] = np.nan

df.to_csv("traffic_missing.csv", index=False)
```

```python
britsIMP = BRITSimputation("traffic_missing.csv")
britsIMP.imputation(100)
britsIMP.plot()
```
